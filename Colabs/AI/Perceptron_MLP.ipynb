# Configuração

Import do dataset da Iris e do MLP, utilizando o sklearn

```python
import numpy as np

from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

Carregamento da Iris

```python
iris = load_iris()

arrays = iris.data
```

Separação dos inputs e dos outputs (x e y)

```python
X = arrays[:,:]
X

data = load_iris()
y = data.target
y
```

*   Utilize como divisão de treino e teste a configuração de 70% por 30%. ✅
*   Garanta o embaralhamento dos dados antes de submetê-los ao treinamento da rede neural. ✅ (O train_test_split() embaralha os dados)



```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Combinações com camadas ocultas

## Todas as combinações com 1 camada oculta, de 1 a 100

```python
best_score = 0
best_params = None

for neurons_layer1 in range(1, 101):
      clf = MLPClassifier(
          hidden_layer_sizes=[neurons_layer1],
          activation='logistic',
          max_iter= 5000,
          random_state=42,
          early_stopping=True,
      )

      clf.fit(X_train, y_train)
      score = clf.score(X_test, y_test)

      print(f"{neurons_layer1}, Score: {score}")

      if score > best_score and score != 1.0 and score:
          best_score = score
          best_params = (neurons_layer1)

print("\nBest Parameters:")
print(f"{best_params}, Score: {best_score}")
```

**Aqui podemos ver que com 32 neuronios em uma única camada oculta, consegue atingir um valor razoável e melhor que quase 70 dos neuronios restantes, mostrando que nem sempre a quantidade ajuda.**

```python
best_score = 0
best_params = None

for neurons_layer1 in range(1, 101):
  clf = MLPClassifier(
    hidden_layer_sizes=[neurons_layer1],
    activation='logistic',
    max_iter= 5000,
    alpha=0.001,
    batch_size=50,
    learning_rate_init=0.01,
    random_state=0,
    early_stopping=True,
      )

  clf.fit(X_train, y_train)
  score = clf.score(X_test, y_test)

  print(f"{neurons_layer1}, Score: {score}")

  if score > best_score and score != 1.0 and score:
      best_score = score
      best_params = (neurons_layer1)

print("\nBest Parameters:")
print(f"{best_params}, Score: {best_score}")
```

**Com os parâmetros diferentes, conseguimos o um resultado melhor com uma quantidade menor de neurônios, sendo mais eficiente**

##Todas as combinações com 2 camadas ocultas, de 1 a 100

```python
best_score = 0
best_params = None

for neurons_layer1 in range(1, 101):
    for neurons_layer2 in range(1, 101):
        clf = MLPClassifier(
            hidden_layer_sizes=[neurons_layer1, neurons_layer2],
            activation='logistic',
            max_iter= 5000,
            random_state=42,
            early_stopping=True,
        )

        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        print(f"{neurons_layer1} : {neurons_layer2}, Score: {score}")

        if score > best_score and score != 1.0:
            best_score = score
            best_params = (neurons_layer1, neurons_layer2)

print("\nBest Parameters:")
print(f"{best_params[0]} : {best_params[1]}, Score: {best_score}")
```

**Esse códgo demora a sua execução por quanta da quantidade de combinações, porém podemos ver que a IA não precisa de mais de 50 neuronios em sua camada, indicando que não precisa de muito para que ela seja eficiente.**

```python
best_score = 0
best_params = None

for neurons_layer1 in range(1, 21):
    for neurons_layer2 in range(1, 21):
        clf = MLPClassifier(
            hidden_layer_sizes=[neurons_layer1, neurons_layer2],
            activation='logistic',
            max_iter= 5000,
            alpha=0.001,
            batch_size=50,
            learning_rate_init=0.01,
            random_state=0,
            early_stopping=True,
        )

        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        print(f"{neurons_layer1} : {neurons_layer2}, Score: {score}")

        if score > best_score and score != 1.0:
            best_score = score
            best_params = (neurons_layer1, neurons_layer2)

print("\nBest Parameters:")
print(f"{best_params[0]} : {best_params[1]}, Score: {best_score}")
```

**Ao rodar esse código com 20:20 e algumas alterções, é possível ver que atingimos um resultado próximo ao anterior, com uma quantidade muito menor de neuronios em cada camada**

## Todas as combinações com 3 camadas ocultas, de 1 a 20

```python
best_score = 0
best_params = None

for neurons_layer1 in range(1, 21):
    for neurons_layer2 in range(1, 21):
      for neurons_layer3 in range(1, 21):
        clf = MLPClassifier(
            hidden_layer_sizes=[neurons_layer1, neurons_layer2, neurons_layer3],
            activation='logistic',
            max_iter= 5000,
            random_state=42,
            early_stopping=True,
        )

        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        print(f"{neurons_layer1} : {neurons_layer2}: {neurons_layer3}, Score: {score}")

        if score > best_score and score != 1.0 and score:
            best_score = score
            best_params = (neurons_layer1, neurons_layer2, neurons_layer3)

print("\nBest Parameters:")
print(f"{best_params[0]} : {best_params[1]} : {best_params[2]}, Score: {best_score}")
```

**Foi utilizado uma quantidade maior de camadas, com menos quantidade de neurônios, podemos ver que utilizando uma baixa quantidade, também chegamos a um resultado próximo a quantidade de valores utilizando apenas 1 camada, que foi 32 neurônios**
